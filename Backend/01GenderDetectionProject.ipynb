# Detecting Face and  saving croped images

```python
haar = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
def extract_images(path,gender,i):
    img= cv2.imread(path)
    gray= cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces=haar.detectMultiScale(gray,1.5,5)
    #scaleFactor : Parameter specifying how much the image size is reduced at each image scale
    #minNeighbors : Parameter specifying how many neighbors each candidate rectangle should have to retain it.
    
    for x,y,w,h in faces:
        roi=img[y:y+h,x:x+w]
        
        if gender == 'male':
            cv2.imwrite('crop/male_crop/{}_{}.png'.format(gender,i),roi)
        else:
            cv2.imwrite('crop/female_crop/{}_{}.png'.format(gender,i),roi)
```

```python
for i,path in enumerate(femalepath):
    try:
        extract_images(path,'female',i)
        print('INFO: {}/{} processesed successfully'.format(i,len(femalepath)))
    except:
        print('INFO: {}/{} unsuccessfull processing'.format(i,len(femalepath)))
```

```python
for i,path in enumerate(malepath):
    try:
        extract_images(path,'male',i)
        print('INFO: {}/{} processesed successfully'.format(i,len(malepath)))
    except:
        print('INFO: {}/{} unsuccessfull processing'.format(i,len(malepath)))
```
